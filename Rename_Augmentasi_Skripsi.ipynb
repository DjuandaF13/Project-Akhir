```python

def rename_images_in_folder(folder_path):
    # Ambil daftar file dalam folder
    files = os.listdir(r"D:\Penelitian Skripsi\Dataset\Gambar Negatif\Tomat")

    
    # Filter hanya gambar dengan ekstensi yang valid
    valid_extensions = ['.jpg', '.jpeg', '.png']
    image_files = [f for f in files if any(f.endswith(ext) for ext in valid_extensions)]
    
    # Loop melalui semua gambar dan ganti nama
    for index, image_name in enumerate(image_files):
        old_path = os.path.join(folder_path, image_name)
        new_name = f"Tomat {index + 1:03d}.jpg"  
        new_path = os.path.join(folder_path, new_name)
        
        # Rename gambar
        os.rename(old_path, new_path)
        print(f"Renamed {image_name} to {new_name}")

# Path folder tempat gambar berada
folder_path = (r"D:\Penelitian Skripsi\Dataset\Gambar Negatif\Tomat")
# Ganti dengan path folder Anda
rename_images_in_folder(folder_path)
```

```python
import cv2  
import os  
  
# Tentukan path folder input dan output  
input_folder = r"D:\Penelitian Skripsi\Dataset\Cabbage detection.v2i.voc\Test Normalisasi"  # Ganti dengan path folder yang berisi gambar asli  
output_folder = r"D:\Penelitian Skripsi\Dataset\Cabbage detection.v2i.voc\Test Normalisasi Hasil"  # Ganti dengan path folder untuk menyimpan gambar yang dinormalisasi  
  
# Buat folder output jika belum ada  
if not os.path.exists(output_folder):  
    os.makedirs(output_folder)  
  
# Loop melalui semua file dalam folder input  
for filename in os.listdir(input_folder):  
    # Cek apakah file adalah gambar  
    if filename.endswith('.jpg') or filename.endswith('.png') or filename.endswith('.jpeg'):  
        # Baca gambar  
        image_path = os.path.join(input_folder, filename)  
        image = cv2.imread(image_path)  
  
        # Normalisasi piksel  
        normalized_image = image / 255.0  
  
        # Simpan gambar yang dinormalisasi ke folder output  
        output_path = os.path.join(output_folder, filename)  
        cv2.imwrite(output_path, normalized_image * 255)  # Mengalikan kembali dengan 255 untuk menyimpan dalam format yang benar  
  
        print(f'Gambar {filename} telah dinormalisasi dan disimpan di {output_folder}')  
  
print('Proses normalisasi selesai.')  
```

```python
import os  #positive.txt
import xml.etree.ElementTree as ET  
  
# Path ke direktori gambar dan file XML  
directory = r"D:\Penelitian Skripsi\Dataset\Cabbage detection.v2i.voc\test"  # Ganti dengan path ke folder yang berisi gambar dan file XML  
  
# Membuat file positives.txt  
output_file = 'positives.txt'  
with open(output_file, 'w') as f:  
    for xml_file in os.listdir(directory):  
        if xml_file.endswith('.xml'):  # Pastikan hanya mengambil file XML  
            # Membaca file XML  
            tree = ET.parse(os.path.join(directory, xml_file))  
            root = tree.getroot()  
  
            # Mengambil informasi dari file XML  
            filename = root.find('filename').text  
            bndbox = root.find('.//bndbox')  
            xmin = bndbox.find('xmin').text  
            ymin = bndbox.find('ymin').text  
            xmax = bndbox.find('xmax').text  
            ymax = bndbox.find('ymax').text  
  
            # Menulis informasi ke dalam file positives.txt  
            f.write(f"{os.path.join(directory, filename)} 1 {xmin} {ymin} {xmax} {ymax}\n")  
            print(f"Processed: {filename}")  # Debugging: Tampilkan nama file yang diproses  
  
print(f"File {output_file} created successfully in {os.getcwd()}.")  # Menampilkan lokasi file yang dibuat  
```

```python
import os  
  
# Path ke direktori yang berisi gambar negatif  
negative_images_path = r"D:\Penelitian Skripsi\Dataset\Gambar Negatif"   # Ganti dengan path ke folder yang berisi gambar negatif  
  
# Membuat file negatives.txt  
with open('negatives.txt', 'w') as f:  
    for img in os.listdir(negative_images_path):  
        if img.endswith('.jpg') or img.endswith('.png'):  # Tambahkan ekstensi lain jika diperlukan  
            f.write(f"{os.path.join(negative_images_path, img)}\n")  
  
print("File negatives.txt telah dibuat.")  
```

```python
import cv2  
import os  
  
# Path ke file positives.txt  
positives_file = 'positives.txt'  # Ganti dengan path ke file positives.txt  
output_vec_file = 'positives.vec'  
  
# Membaca file positives.txt  
with open(positives_file, 'r') as f:  
    lines = f.readlines()  
  
# Membuat file vektor  
samples = []  
for line in lines:  
    parts = line.strip().split()  
    img_path = parts[0]  
    samples.append(cv2.imread(img_path))  
  
# Menyimpan file vektor  
cv2.imwrite(output_vec_file, samples)  
  
print(f"File vektor '{output_vec_file}' telah dibuat.")  
```
